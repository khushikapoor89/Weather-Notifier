```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# OpenWeatherMap API key (replace with your actual API key)
API_KEY = "788bd6dc724ecca8c558fa0ff4bb0c92"
```

```python
# Function to fetch current weather data
def get_weather(city_name):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={API_KEY}&units=metric"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        return None
```

```python
# Function to display weather details
def display_weather(data):
    city = data['name']
    temp = data['main']['temp']
    humidity = data['main']['humidity']
    weather_desc = data['weather'][0]['description']

    print(f"City: {city}")
    print(f"Temperature: {temp}°C")
    print(f"Humidity: {humidity}%")
    print(f"Weather Description: {weather_desc}")
```

```python
# Function to plot weather trends
def plot_weather_trends(history):
    df = pd.DataFrame(history)

    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x='Timestamp', y='Temperature', marker="o",markersize=15, label='Temperature (°C)',color='red')
    sns.lineplot(data=df, x='Timestamp', y='Humidity', marker="s",markersize=15, label='Humidity (%)',color='green')

    plt.title('Weather Trends')
    plt.xlabel('Timestamp')
    plt.ylabel('Values')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
def weather_notifier():
    city = input("Enter the name of the city: ")
    history = []

    while True:
        # Fetch and display weather data
        weather_data = get_weather(city)
        if weather_data:
            display_weather(weather_data)
            # Store data but limit history to last 10 entries
            history.append({
                'Timestamp': pd.Timestamp.now(),
                'Temperature': weather_data['main']['temp'],
                'Humidity': weather_data['main']['humidity']
            })
            if len(history) > 1:  # Keep only the last 10 entries
                history = history[-1]

            # Plot trends if user wants to
            if input("See weather trends so far? (yes/no): ").strip().lower() == 'yes':
                plot_weather_trends(history)
                break
        else:
            print("Could not retrieve weather data. Check city name or internet connection.")
            break
```

```python
# Run the notifier
weather_notifier()
```
